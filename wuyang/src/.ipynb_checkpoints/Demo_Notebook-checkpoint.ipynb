# Deviation of index tracking funds from indexes
* index: market benchmark, e.g. S&P 500
* index tracking fund: try to track the performance of a specific index, e.g. SPDR S&P 500 ETF
    - fund could be leveraged, e.g. The Proshares Ultra S&P500 seeking 200% S&P 500
    - fund could be inversely tracking, e.g. The ProShares UltraPro Short QQQ seeking -300% NASDAQ-100 Index

### Problem: predicting whether fund tracks the index well or not
- The criteria is to see whether the fund moves the amount that they should do.

### Why this problem?
- The index tracking funds have huge assets, such as SPDR S&P 500 has $170+B and 
    there are also many other funds with multi-billion assets, so that the action
    of these funds may lead to the movement of specific stock or even the market.
- The deviation of the funds from indexes would indicate an action in the near future
    by funds and this would be a good market signal and helps construct a trading strategy.

###We use one index-fund pair as our example.

[The Nasdaq 100 index](http://www.nasdaq.com/markets/indices/nasdaq-100.aspx): The NASDAQ-100 Index includes 100 of the largest domestic and international non-financial securities listed on The Nasdaq Stock Market based on market capitalization.

[ProShares UltraPro Short QQQ](http://www.proshares.com/funds/sqqq.html):ProShares UltraPro Short QQQ seeks daily investment results, before fees and expenses, that correspond to three times the inverse (-3x) of the daily performance of the NASDAQ-100 Index.

```python
index_code = "YAHOO/INDEX_NDX"
fund_code = "GOOG/NYSE_SQQQ"
direction = -3
start_date = "2005-01-01"
```

** We collect the data from [Quandl](http://www.quandl.com/) through their api **

```python
from fund_index_pair import fund_index_pair # class that holds the data with member functions processing the data
fund_and_index = fund_index_pair(fund_code, index_code, direction, start_date) # initialize one instance
fund_and_index.load()
fund_and_index.data.columns # show the information columns
```

```python
fund_and_index.data[index_code+'_Adjusted Close'].plot() #plot the adjusted close price which is the tracking target of the fund
plt.show()
```

### remove outliers

```python
fund_and_index.remove_outliers_by_moving_average()
fund_and_index.data[index_code+'_Adjusted Close'].plot()
plt.show()
```

###Extend features according to intuitions

**The extended features include**: 

1. ***inner day change*** as (open-close)/open, 
2. ***daily change*** as (today's close - yesterday's close)/(yesterday's close), 
3. ***next day open move*** as (next day's open - today's close)/(today's close),
4. ***daily volatility*** as (daily high - daily low)/(daily open)
5. ***deviatioin of the index movement from its target movement (with respect to index movment)***

We could include the ***next day open move*** because we are predicting the deviation at close price movements in the morning of the day. We would like the market to absorb the news released overnight.

**Trarget varialbe**:

***Next day deviation*** or ***absolute value of next day's deviation***

```python
fund_and_index.extend_data() # extend features
fund_and_index.data.info() # show the information
```

### Feature selection

We only select features that are somewhat correlated with target variable. There could be many other feature selection methods, e.g. PCA. We would like to keep the original features so that we coudl identify their importance at the end.

```python
# correlation coefficient matrix
mat_corr = fund_and_index.data.dropna().corr()
# drop the correlation of predicting variables
mat_corr = mat_corr[(mat_corr.index != 'next_day_diff') & (mat_corr.index != 'next_day_abs_diff')]
figure = mat_corr[['next_day_diff', 'next_day_abs_diff']].plot(grid=True,use_index=True)
plt.xticks(rotation=45)
plt.show()
```

```python
important_feature_list = mat_corr.loc[mat_corr['next_day_diff'].abs() > 0.05].index.tolist()
print important_feature_list
```

### Form features by keeping only the important columns of data
**The target variable could be absolute value instead of directional value.
Here, we use the directional values as an example. These value will later be thresholded to make the problem a classification problem, which could be relatively easier.**

```python
features = fund_and_index.data.dropna()[important_feature_list]
target_variable = fund_and_index.data.dropna()['next_day_diff']
```

### histogram of the normal 95% of target variable (deviations)

```python
# Interactive histogram: distribution of the target variable
# More detailed distribution could be viewed by changing the head and tail cutoff values
from IPython.html.widgets import interact
def interact_histogram(head, tail):
    target_variable[(target_variable < np.percentile(target_variable, tail)) & (target_variable > np.percentile(target_variable, head))].hist(bins=200)
    plt.show()
interact(interact_histogram, head = (0,50), tail=(50,100))
```

## Regression
Now we could explore regressions to see whether it is possible to predict the real deviation value

To make the prediction, we need to train a model (linear here) with data collected over a certain amount of time (***history_days***). Then we use this model to make predictions in the future. The model is only for predicting the deviation of next days given all infomation of today (the selected features). The ***future_days*** parameter in the following cell means how long we use the current model in the future, until we re-train the model. ***future_days***=1 means we re-train model every day and use that model to predict the next day's deviation. ***n_split*** is how many times we do this model training and prediction. The average prediction accuracies/errors will be reported.

```python
future_days = 1 # the future days to apply a fixed trained model
history_days = 100 # the past history to train a model
n_split = 500 # the total moving window
```

```python
# The regression_bundle contains functions that put several linear regression models into a group 
# and apply all of them on the input data so that we could compare the accuracies of these models.
# This is a single function at this point and will be reformatted into a class with member functions that
# add(delete) models into(from) the bundle
from regression_bundle import multiTimeSplit_models_train_and_eval
regress_accus = multiTimeSplit_models_train_and_eval(features, target_variable, future_days, history_days, n_split)
regress_accus.mean()
```

```python
# whos, then delete the results of linear regression to free space
del regress_accus
```

## Observation:
**Regression is only a little better than Baseline (always predict the deviation of the next day by the average deviation in the training samples)**
###Transform the problem to a classification problem by thresholding the target variable

```python
n_split = 100 # the total moving window, we use 100 instead of 500 due to time constraints
percentile = 50 # threshold to create labels
classification_labels = (target_variable) > np.percentile(target_variable[-n_split*future_days-1:-1], percentile)
```

```python
# Similar to the regression case, the classification bundle is a collection of multiple classifiers.
# The function in this bundle will apply each classifier to the input data and output the prediction accuracies and the tuned models.
# Model parameters of each classifier will be tuned through cross-validation on the training samples and the optimal model will be
# applied to the test samples. There is a baseline classifier that we applied for comparison.
# The baseline classifier predict the label of test sample by the majority label of the training samples.
from classification_bundle import multiTimeSplit_classifier_eval
classification_accus, classification_models = multiTimeSplit_classifier_eval(features, classification_labels, future_days, history_days, n_split)
```

```python
classification_accus.mean().plot(grid=True, use_index=True)
plt.xticks(rotation=45)
plt.show()
```

## Observation:
**Every method is better than baseline and linear SVM is the best one with accuracy ~68%**
### We next examine what is inside Linear SVM
For each linear SVM classifier, there are weights for each feature. We try to show these weights over time.

```python
linearSVM_abs_coefs = []
for i in range(len(classification_models)):
    linearSVM_abs_coefs.append(np.abs(classification_models[i][1].coef_[0])) # use absolute coefficients as importance factors
linearSVM_abs_coefs = pd.DataFrame(linearSVM_abs_coefs, columns=important_feature_list)
```

```python
figure = linearSVM_abs_coefs.plot()
figure.legend(bbox_to_anchor = (1.7, 1))
plt.show()
linearSVM_abs_coefs.mean().plot(use_index=True)
plt.xticks(rotation=45)
plt.show()
```

```python
# remove resulting variables to free space
del classification_accus, classification_labels, classification_models, linearSVM_abs_coefs 
```

## Observation:
1. The models vary a lot across time, reflecting the movement of the market.
2. The most important features are:
    1. *onDay_index_openMove*: the movement of the open price in the prediction day from the close price of the previous day.
    2. *diff_changes*: the deviation of the fund movement from its tracking target induced by index movement.
These important features are quite intuitive. The *onDay_index_openMove* is the most recent information we collect before prediction.
On the other hand, the *diff_changes* is essentially the label of sample before prediction. 

## Other explorations:
we could create labels in various other ways to fit different goals. Here is one more example.
Instead of predicting the sign of the deviation of fund's move to its target move, we could predict whether the deviabtioin (absolute value) is large or not.


```python
# we need to re-define the target variable and re-select the features
important_feature_list = mat_corr.loc[mat_corr['next_day_abs_diff'].abs() > 0.05].index.tolist()
print important_feature_list

features = fund_and_index.data.dropna()[important_feature_list]
target_variable = fund_and_index.data.dropna()['next_day_abs_diff']
```

```python
# recreate labels as whether the absolute deviation is large (> average devation) or small
# since the target_variable is already the absolute deviation, we don't need to take absolute operation again
classification_labels = (target_variable) > np.percentile(target_variable[-n_split*future_days-1:-1], percentile)
```

```python
time1 = time.time()
classification_accus, classification_models = multiTimeSplit_classifier_eval(features, classification_labels, future_days, history_days, n_split)
print time.time()-time1
classification_accus.mean().plot(grid=True, use_index=True)
plt.xticks(rotation=45)
plt.show()
```

## Observation:
1. From the prediction accuracies, this prediction of high deviation is a harder problem than predicting the sign of deviation.
2. Among all methods, the logistic regression method performs the best.

**Next we will further check the magnitudes of the parameters in logistic regression model**

```python
logitRegress_abs_coefs = []
for i in range(len(classification_models)):
    logitRegress_abs_coefs.append(np.abs(classification_models[i][1].coef_[0])) # use absolute coefficients as importance factors
logitRegress_abs_coefs = pd.DataFrame(logitRegress_abs_coefs, columns=important_feature_list)
```

```python
figure = logitRegress_abs_coefs.plot()
figure.legend(bbox_to_anchor = (1.7, 1))
plt.show()
logitRegress_abs_coefs.mean().plot(use_index=True)
plt.xticks(rotation=45)
plt.show()
```
