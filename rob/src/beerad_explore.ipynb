```python
%%sql
select br.name, count(be.id) as beer_ct
from brewers br inner join beers be
    on br.id = be.brewer_id
group by br.name
order by beer_ct desc
limit 20
```

```python
%%sql
select br.location, count(be.id) as beer_ct
from brewers br inner join beers be
    on br.id = be.brewer_id
group by br.location
order by beer_ct desc
limit 20
```

```python
%%sql
select st.name, count(be.id) as beer_ct
from styles st inner join beers be
    on st.id = be.style_id
group by st.name
order by beer_ct desc
limit 20
```

```python
%%sql
select br.location, st.name, count(be.id) as beer_ct, tot as tot_beers, count(be.id)/tot as proportion
from 
    styles st inner join beers be
        on st.id = be.style_id
    inner join brewers br
        on be.brewer_id = br.id
    inner join (
        select bri.location, count(bei.id) as tot
        from brewers bri inner join beers bei
            on bri.id = bei.brewer_id
        group by bri.location
        order by tot desc
        limit 20
    ) tot
        on tot.location = br.location
group by st.name, tot, br.location
having count(be.id)/tot_beers > 0.05
order by tot_beers desc, beer_ct desc
limit 20
```

```python
%%sql
select st.name, count(distinct be.id), tot_beer_ct
from 
    (   select * 
        from styles 
        where name like '%ale%' or name like '%ipa%'
    ) st inner join
    (   select * 
        from beers
        where ibu > 0
    ) be 
        on be.style_id = st.id 
    inner join (   
        select st.id, count(be.id) as tot_beer_ct
        from styles st inner join beers be
            on st.id = be.style_id
        group by st.name 
    ) bet
        on bet.id = st.id
        
group by st.name, tot_beer_ct
```
