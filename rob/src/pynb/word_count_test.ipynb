```python
from nltk.corpus import movie_reviews
from nltk.corpus import brown

news_text = brown.words(categories='reviews')
good_movie_revs = movie_reviews.words(categories='pos')
bad_movie_revs = movie_reviews.words(categories='neg')

word_list = [w.lower() for w in news_text]
word_list.extend([w.lower() for w in good_movie_revs])
word_list.extend([w.lower() for w in bad_movie_revs])

fdist = nltk.FreqDist([w.lower() for w in word_list])
modals = ['can', 'could', 'may', 'might', 'must', 'will', 'hop', 'beer', 'flavor']
for m in modals:
    print m + ':', fdist[m],
```
