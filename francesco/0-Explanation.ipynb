##The dataset

I am using datasets from Yelp, consisting of about 600,000 reviews.
The test set, which contains 23,000 reviews, is not used during feature selection and model training.

###Metrics

- Accuracy (or number of correctly identified ratings)
- Mean absolute and squared error
- One-away ratio (reviews predicted at most 1 away from the ground truth)

###First models

My first quick classifier was a Naive Bayes on bag of words, selected by frequency. Accuracy between 50 and 54%, doesn't improve much increasing size of the bag of words.

##Negations
Misclassified reviews revealed the need of better treatmen of negations.
I decided to add a preprocessing step and attach every occurrence of `not` to the following word.
For example,

`not great`

was transformed into

`not_great`



##N-grams

To get past the accuracy wall of the previous method, I employed higher order n-grams. This time, instead of selecting by frequency across the entire corpus, I subdivided the reviews into three categories:

- negative (1 or 2 stars)
- neutral (3 stars)
- positive (4 or 5 stars)

and selected frequent N-grams in each category. I then merged the sets of n-grams before feeding them to the vectorizer.
(I did this because the training set contains relatively few negative reviews; as a consequence selecting by frequency would yield a larger amount of features from positive reviews. This reflected in the results of early classifiers, where accuracy on negative reviews was significantly lower than that on positive ones.)

##Results

Here are the results for Naive Bayes and Logistic Regression classifiers.
Logistic regression performs better on all fronts, achieving an accuracy of 63.7%, and predicting at within 1 from the ground truth in 93.8% of the cases. Curiously Logistic Regression tends to overshoot compared to the naive Bayes classifier.

```python
# load classifiers...
vectorizer, nbclassifier = classifiers.load('naivebayes')
lrclassifier = classifiers.load_classifier('logistic')
testdf = test_set()

X = vectorizer.transform(testdf.text) # takes a few seconds
y = testdf.stars
```

```python
X = vectorizer.transform(testdf.text) # takes a few seconds
y = testdf.stars.as_matrix()

nbpredictions = nbclassifier.predict(X)            # predictions according to naive Bayes
lrpredictions = lrclassifier.predict(X)            # predictions according to logistic regression
mfpredictions = np.zeros_like(lrpredictions) + 4   # always predict 4 - most frequent rating in training set
```

```python
print("Accuracy for naive Bayes on the test set: %.2f%%" % (nbclassifier.score(X, y)*100))
print("Accuracy for logistic regression classifier: %.2f%%" % (lrclassifier.score(X, y)*100))
```

```python
print("Absolute error, naive Bayes: %s" % mean_abs_error(y, nbpredictions))
print("Absolute error, logistic regression: %s" % mean_abs_error(y, lrpredictions))
```

```python
print("One-away percentage, naive Bayes: %.2f%%" % (one_away_ratio(y, nbpredictions)*100))
print("One-away percentage, logistic regression: %.2f%%" % (one_away_ratio(y, lrpredictions)*100))
```
