```python
#get an entire years data for a given city
from datetime import date
import json,urllib2
yr= 2014
sMonth = 1
eMonth = 5
city = "Baltimore"
yrData = []

for  cnt in range(sMonth, eMonth + 1):
    startDt = '%20' + '%i' %yr + '-%i' %cnt + '-01'
    if cnt<12:
        endDt = '%20' + '%i' %yr + '-%i' %cnt + '-%i' %(date(yr, cnt+1, 1) - date(yr, cnt, 1)).days
    elif cnt == 12:
        endDt = '%20' + '%i' %yr + '-%i' %cnt + '-%i' %31
    url = 'https://api.worldweatheronline.com/premium/v1/past-weather.ashx?q=%s&format=json&date=%s&enddate=%s&tp=24&key=bd5fb86dc432f72d06dc8a167648b23db24a9258' % (urllib2.quote(city), startDt,endDt) 
    #print url
    ret = urllib2.urlopen(url).read()
    monthData = json.loads(ret)
    yrData = concatenate( (yrData, monthData['data']['weather'] ))
    print "completed: " + startDt[3:] + ' through '+ endDt[3:]
```
